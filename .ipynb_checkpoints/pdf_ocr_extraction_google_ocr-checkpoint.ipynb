```python
import pymupdf # import PyMuPDF
import glob
import os
```

```python
cls = 'class_10'
sub = 'social'
folder = "prev_year_papers"
```

```python
search_directory = "./backend/ncert/"+cls+"/"+sub+"/"+folder+"/"  # Replace with your directory path
file_format = '*.pdf'  # Replace with the file format you're looking for
```

```python
list_pdfs=glob.glob(os.path.join(search_directory, '**', file_format), recursive=True)
```

```python
list_pdfs
```

```python
for file_no in range(len(list_pdfs)):
    file=list_pdfs[file_no]
    print(file)
    file_txt=file.replace("pdf","txt")
    doc = pymupdf.open(file)
    all_text = chr(12).join([page.get_text(sort=True) for i, page in enumerate(doc, start=1) if i >= 3 and i % 2 != 0])
    file2write=open(file_txt,'w',encoding="utf-8")
    file2write.write(all_text)
    file2write.close()
```
